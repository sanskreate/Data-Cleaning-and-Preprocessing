#Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

#Data Cleaning and Data Preprocessing

```python
df = pd.read_csv("marketing_campaign.csv", sep="\t")
df
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
#cleaning column titles
def clean_column_names(df):
  cleaned_names = []
  for col in df.columns:
    col = col.replace(' ', '_')
    col = col.lower()
    cleaned_names.append(col)

  df.columns = cleaned_names
  return df

df = clean_column_names(df)
df.head()
```

```python
#converting to datetime
df = df.copy()
df["dt_customer"] = pd.to_datetime(df["dt_customer"], dayfirst=True)
first_date = df["dt_customer"].min()
last_date = df["dt_customer"].max()

print("First date:", first_date)
print("Last date:", last_date)
```

```python
#Feature engineering
df = df.copy()

# 1. Total Spend
df['total_spend'] = df['mntwines'] + df['mntfruits'] + df['mntmeatproducts'] + df['mntfishproducts'] + df['mntsweetproducts'] + df['mntgoldprods']

# 2. Total Children
df['total_children'] = df['kidhome'] + df['teenhome']

# 3. Age
current_year = pd.to_datetime('today').year
df['age'] = current_year - df['year_birth']

# 4. Customer Seniority (in years)
df['customer_seniority'] = (pd.to_datetime('today') - df['dt_customer']).dt.days / 365.25

# 5. Total Purchases
df["total_purchases"] = df["numwebpurchases"] + df["numcatalogpurchases"] + df["numstorepurchases"]

# 6. Campaign Acceptance Rate
df['campaign_acceptance_rate'] = (df['acceptedcmp1'] + df['acceptedcmp2'] + df['acceptedcmp3'] + df['acceptedcmp4'] + df['acceptedcmp5'] + df['response']) / 6

# 7. Family Size
df['family_size'] = df['total_children'] + 2 # Adding 2 for parents

# creating bins for age group
bins = [18, 30, 45, 60, 100]
labels = ['18-29', '30-44', '45-59', '60+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)


df.head()
```

```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
print("Categorical variables:\n", categorical_cols.tolist())
print("\nUnique values per categorical column:")
for col in categorical_cols:
    print(f"\nColumn: {col}")
    print(df[col].unique())
```

```python
#label encoding categorical cols
LE = LabelEncoder()
for col in categorical_cols:
    df[col] = LE.fit_transform(df[col])

print("All categorical features have been label encoded.")
```

```python
df.describe()
```

```python
#checking cols for outlier detection
cols_to_plot = ['income', 'age', 'total_spend']
sns.pairplot(df[cols_to_plot])
plt.suptitle("Pairplot for Outlier Detection", y=1.02)
plt.show()
```

```python
def remove_outliers_iqr(data, columns):
    clean_data = data.copy()
    for col in columns:
        Q1 = clean_data[col].quantile(0.25)
        Q3 = clean_data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        clean_data = clean_data[(clean_data[col] >= lower_bound) & (clean_data[col] <= upper_bound)]
    return clean_data

numerical_cols = ['income', 'age', 'total_spend']
df_clean = remove_outliers_iqr(df, numerical_cols)

print("Original shape:", df.shape)
print("Shape after outlier removal:", df_clean.shape)
```

```python
def compare_distributions(original_df, cleaned_df, columns):
    for col in columns:
        fig, axes = plt.subplots(1, 2, figsize=(12, 5))

        # Boxplot comparison
        sns.boxplot(y=original_df[col], ax=axes[0], color='salmon')
        axes[0].set_title(f'Before Outlier Removal - {col}')

        sns.boxplot(y=cleaned_df[col], ax=axes[1], color='lightgreen')
        axes[1].set_title(f'After Outlier Removal - {col}')

        plt.tight_layout()
        plt.show()

        # Histogram comparison
        fig, axes = plt.subplots(1, 2, figsize=(12, 5))

        sns.histplot(original_df[col], bins=30, kde=True, ax=axes[0], color='salmon')
        axes[0].set_title(f'Histogram Before - {col}')

        sns.histplot(cleaned_df[col], bins=30, kde=True, ax=axes[1], color='lightgreen')
        axes[1].set_title(f'Histogram After - {col}')

        plt.tight_layout()
        plt.show()

compare_distributions(df, df_clean, numerical_cols)
```
